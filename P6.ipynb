```python
from tensorflow.keras.utils import load_img
from tensorflow.keras.utils import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16
```

```python
image = load_img(r'C:\Users\Admin\Downloads\dog.jpg', target_size=(224, 224))
```

```python
#convert image pixels to nuumpy array
image = img_to_array(image)
```

```python
print(image)
```

```python
#reshape data for the model
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
```

```python
print(image)
```

```python
#prepare the process for VGG model
image = preprocess_input(image)
```

```python
#Load the model
model = VGG16()
```

```python
#printed array represents the pixel values of the image after preprocessing
print(image)
```

```python
#predicts the probability across all output classs
#yhat variable will contain the model's prediction output based on the input image. 
yhat =model.predict(image)
```

```python
#convert the probabilities to class labels
label=decode_predictions(yhat)
```

```python
#retrive the most likely result eg. highest probability
label=label[0][0]
```

```python
#print the classification
print('%s (%.2f%%)' % (label[1], label[2]*100))
```
